```python
#Mã hóa
def LABEL_ENCODING(c1):
    from sklearn import preprocessing
    label_encoder = preprocessing.LabelEncoder()
    data[c1]= label_encoder.fit_transform(data[c1])
    data[c1].unique()
LABEL_ENCODING("weather")
data
```

```python
#Loại bỏ cột date
data = data.drop('date',axis=1)
```

```python
x = data.drop('weather',axis=1)
y = data['weather']
```

```python
#Chia tách tập dữ liệu thành tập huấn luyện và tập kiểm tra.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Tạo và huấn luyện một mô hình Decision Tree Classifier.
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
# Dự đoán các giá trị của biến y cho tập dữ liệu thử nghiệm X_test
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#Biểu đồ nhiệt
sns.heatmap(cm,annot=True)
```

```python
# Tính toán độ chính xác của mô hình Decision Tree Classifier
acc1 = accuracy_score(y_test, y_pred)
print(f"Accuracy score: {acc1}")
```

```python
mylist=[]
mylist2=[]

mylist.append(acc1)
mylist2.append("DTR")

plt.rcParams['figure.figsize']=8,6
sns.set_style("darkgrid")
plt.figure(figsize=(8,8))
ax = sns.barplot(x=mylist2, y=mylist, palette = "mako", saturation =1.5)
plt.xlabel("Classification Models", fontsize = 20 )
plt.ylabel("Accuracy", fontsize = 20)
plt.title("Accuracy of Classification Model", fontsize = 20)
plt.xticks(fontsize = 11, horizontalalignment = 'center', rotation = 8)
plt.yticks(fontsize = 13)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.2%}', (x + width/2, y + height*1.02), ha='center', fontsize = 'x-large')
plt.show()
```

```python
# Tạo mảng NumPy chứa các tính năng đầu vào cho dự báo thời tiết.
weather_features = np.array([[0, 5.6, -2.1, 4.7]])


# Tạo đối tượng DecisionTreeClassifier.
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
# Dự báo thời tiết ngày mai.
weather_prediction = classifier.predict(weather_features)
# Hiển thị thời tiết dự báo
print(f"The predicted weather for tomorrow is: {weather_prediction}")
if(weather_prediction==0):
    print("Drizzle")
elif(weather_prediction==1):
    print("Fog")
elif(weather_prediction==2):
    print("Rain")
elif(weather_prediction==3):
    print("snow")
else:
    print("Sun")

```
